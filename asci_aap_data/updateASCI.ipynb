```python
from selenium import webdriver
from selenium.webdriver.common.by import By # contains operators for the type of search we want to do
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import csv
```

## I am looking at 2022

```python
#Looking at 2022
driver = webdriver.Chrome()
driver.get('https://the-asci.org/controllers/asci/ListingController.php?action=year&key=2022')
```

```python
numPeople = 0
allindividuals = []
parent_element = driver.find_element('xpath', '//*[@id="members-elected-for-year"]/tbody')
sub_categories = parent_element.find_elements('xpath', './/tr/td[2]/a[@href]')
for sub_category in sub_categories:
    href = sub_category.get_attribute('href')
    print(href)
    allindividuals.append(href)
    numPeople += 1
print(f'Number of people in year 2022: {numPeople}')

```

```python
allNames=[]
for link in allindividuals:
    indvInfo = []
    
    try:
        driver.get(link)  

        #Find the first name and last name
        div_element = driver.find_element(By.CLASS_NAME, 'header')

        # Extract and print the text content of the div
        text_content2 = div_element.text
        name_without_credentials = text_content2.split(',')[0]
        edit_name = name_without_credentials.split()
        first_name = edit_name[0]
        last_name = edit_name[-1]

        indvInfo.append(first_name)
        indvInfo.append(last_name)

        print(f'"Name: {first_name} {last_name}')

         # Find the affiliation
        affil = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/div/div[4]/div[2]/a')
        text_content1 = affil.text
        print(text_content1)

        indvInfo.append(text_content1)


         #Find the concepts
        
        speciality = driver.find_element(By.XPATH, '//*[@id="primary"]/div/div[2]/div[3]/div[1]')

        # Find sub-categories underneath the speciality element
        sub_categories = speciality.find_elements(By.XPATH, './/div/a')

        # Initialize a list to store the href values
        all_Concepts = []

        # Iterate through the sub-categories and print or process them as needed
        for sub_category in sub_categories:
            text_content3 = sub_category.text
            print(text_content3)
            all_Concepts.append(text_content3)

        indvInfo.append(all_Concepts)

        allNames.append(indvInfo)     
        
    except Exception as e:
        
        print(f"Error processing {link}: {str(e)}")
                                  
    
```
